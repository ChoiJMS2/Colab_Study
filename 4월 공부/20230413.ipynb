```python
def hello(name:str) -> None:
  if not isinstance(name, str):
    raise TypeError("Name Must be a string!")
  print(f"안녕!,{name}")
```

```python
def add(a, b, c):
  return a+b+c

print(add(1,2,3))
```

```python
def add(*args):
  print(args, type(args))

add(1,2,3,4,5)
```

```python
def add(*args):
  addition = 0
  for number in args:
    addition += number
  return addition

print(add(1,2))
print(add(1,2,3))
print(add(1,2,4,5))
```

```python
# 머신러닝 클래스, 하이퍼파라미터 튜닝 시, 입력값이 딕셔너리를 기본값으로 받음
def temp(**kwargs): #딕셔너리 형태
  for key, value in kwargs.items():
    print(key, value)

temp(name="Choi", age=36, city="Suwon")
```

```python
# 행 추출, ZN = 18.0인 값만 조회하여 CRIM, ZN, INDUS 칼럼 추출
cols = ['CRIM', 'ZN', 'INDUS']
result3 = df_boston.loc[(df_boston["ZN"] == 18.0), cols]
result3
```

```python
## 다중조건식
# result = 데이터.loc[행, 열]
# result = 데이터.loc[ () & () | () , 열 ]
cols = ['CRIM', 'ZN', 'INDUS']

# CRIM이 1보다 작고, target이 24이상인 값만 조회
df_boston.loc[(df_boston['CRIM'] < 1) & (df_boston['target'] >= 24), cols]

# CRIM이 1보다 작거나, target이 24이상인 값만 조회
df_boston.loc[(df_boston['CRIM'] < 1) | (df_boston['target'] >= 24), cols]
```

```python
# rename 컬럼의 이름값 변경 --> 딕셔너리
# ZN --> landZone
result = df_boston.rename(columns={'ZN' : 'landZone'})

tempDict={
    'ZN' : 'landZone',
    'CRIM' : 'A',
    'CHAS' : 'B',
    'TAX' : 'C'
}
result = df_boston.rename(columns = tempDict)
result.head(1)
```

```python
sales.info()
```

```python
sales['date'] = pd.to_datetime(sales['datesold'])
sales.head(1)
```

```python
sales['datesold'] = pd.to_datetime(sales['datesold'])
sales['year'] = sales['datesold'].dt.year
sales['month'] = sales['datesold'].dt.month
sales['day'] = sales['datesold'].dt.day
sales.head(1)
```

```python
sales['datesold'].max() - sales['datesold'].min()
```

### shift()

```python
temp_df = sales[['datesold', 'price']].copy()
temp_df['shifted_v1'] = temp_df['price'].shift(1, fill_value=0).astype(int)
temp_df['shifted_v2'] = temp_df['price'].shift(2, fill_value=0).astype(int)

temp_df['차이값'] = temp_df['price'] - temp_df['shifted_v1']
temp_df.head()
```

```python
# 0000년 00월 00일
temp_df['한글날짜'] = temp_df['datesold'].dt.strftime('%Y년 %m월 %d일')
temp_df.head()
```
